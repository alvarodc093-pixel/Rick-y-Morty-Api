```python
def flatten_character(character):
    """JSON de un personaje a un diccionario plano"""
    status = character["status"]
    type_ = character["type"]
    gender = character["gender"]
    origin = character["origin"]["name"]
    location = character["location"]["name"]
    image = character["image"]
    episode = character["episode"]
    url = character["url"]
    created = character["created"]

    return {
        "id": character["id"],
        "name": character["name"],
        "status": character["status"],
        "species": character["species"],
        "type": character["type"],
        "gender": character["gender"],
        "origin": character["origin"]["name"],
        "location": character["location"]["name"],
        "image": character["image"],
        "episode": character["episode"],
        "url": character["url"],
        "created": character["created"]
    }

flatten_character(first_character)
```

```python
import time

rows = []
page = 1

while True:
    r = requests.get(f"{BASE}?page={page}", timeout=30)
    if r.status_code == 429:
        time.sleep(1)
        continue
    r.raise_for_status()

    data = r.json()
    for character in data["results"]:
        rows.append(flatten_character(character))

    if not data["info"]["next"]:
        break
    page += 1

df = pd.DataFrame(rows)
print("Forma:", df.shape)
df.head()

df.to_csv("rick_and_morty_characters.csv", index=False)
print("CSV guardado en rick_and_morty_characters.csv")
```
